```python
import numpy as np
import random
from matplotlib import pyplot as plt

data = np.random.normal(0, 20, 1000)


bins = np.arange(-100, 100, 5) # fixing bin size

plt.xlim([min(data)-1, max(data)+1])

plt.hist(data, bins=bins, color='blue', alpha=0.5,  edgecolor='Black')
plt.show()
```

```python
plt.hist(data, bins=bins, color='blue', alpha=0.5,  edgecolor='Black')
plt.grid()

plt.show()
```

```python
plt.hist(data, bins=bins,color='blue', alpha=0.5,  edgecolor='Black')
ax = plt.gca()
ax.xaxis.grid()

plt.show()
```

```python
plt.hist(data, bins=bins,color='blue', alpha=0.5,  edgecolor='Black')
ax = plt.gca()
ax.yaxis.grid()

plt.show()
```

```python
from matplotlib import rcParams

rcParams['grid.linestyle'] = '-'
rcParams['grid.color'] = 'black'
rcParams['grid.linewidth'] = 0.1

plt.grid()
plt.hist(data, bins=bins,color='blue', alpha=0.5,  edgecolor='Black')

plt.show()
```

```python
import numpy as np

# major ticks every 20
major_ticks = np.arange(-100, 101, 20)

ax = plt.gca()
ax.yaxis.grid()

ax.set_yticks(major_ticks)

plt.hist(data, bins=bins,color='blue', alpha=0.5,  edgecolor='Black')

plt.show()
```
